## Exercise 2 - Import Matplotlib using




```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
fig, ax = plt.subplots(figsize=(7,4))
ax.set_title('Sin Wave')
ax.plot(x,y)
pltshow()
```

```python
x = np.array([1989, 1992, 1990, 1995, 2000, 2025, 2022, 2018, 2019, 2023])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel=("Year")
plt.ylabel("Price")
plt.show()
```

## Exercise 3 - Draw Barchart




```python
x = np.array(["a", "b", "c", "d"])
y = np.array([5, 7, 3, 11])
plt.bar(x,y)
plt.show()
```

## Exercise 5 -  Draw Piechart



```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Generating 3D data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Creating a basic 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='blue', marker='o', label='Data Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('Basic 3D Scatter Plot')
```
